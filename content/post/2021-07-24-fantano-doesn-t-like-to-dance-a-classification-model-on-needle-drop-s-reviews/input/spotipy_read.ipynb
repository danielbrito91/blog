# Query Spotify

```python
#https://rareloot.medium.com/extracting-spotify-data-on-your-favourite-artist-via-python-d58bc92a4330
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

client_id = #inform your client id
client_secret = #inform yout client secret

client_creds_man = SpotifyClientCredentials(client_id = client_id, client_secret = client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_creds_man)
```

```python
def find_album(artist_search, album_search, album_id):
    album_query = sp.search(q ='artist: ' + artist_search  + ' album:' + album_search, type = 'album')
    if album_query['albums']['items'] == []:
        album_name = None
        album_uri = None
        artist_name = None
    else:
        album_uri = album_query['albums']['items'][0]['uri'] #''spotify:album:1PQDjdBpHPikAodJqjzm6a''
        album_name = album_query['albums']['items'][0]['name']
        album_type = album_query['albums']['items'][0]['type']
        album_popularity = sp.album(album_uri)['popularity']
        album_genres = sp.album(album_uri)['genres']
        artist_name = album_query['albums']['items'][0]['artists'][0]['name']
        artist_uri = album_query['albums']['items'][0]['artists'][0]['uri']
        artist_genres = sp.artist(artist_uri)['genres']
    return(album_id,
           album_name, album_uri,
           album_type, album_popularity, album_genres,
           artist_name, artist_uri, artist_genres)
```

```python
def find_album_uri(artist_search, album_search, alb_id):
    album_query = sp.search(q ='artist: ' + artist_search  + ' album:' + album_search, type = 'album')
    albums_uri_db = {'album_id':[],
                     'album': [],
                     'album_uri': [],
                     'artist': [],
                     'artist_uri':[]}
    
    if album_query['albums']['items'] == []:
        alb_name = None
        alb_uri = None
        art_name = None
        art_uri = None
    else:
        alb_uri = album_query['albums']['items'][0]['uri']
        alb_name = album_query['albums']['items'][0]['name']
        art_name = album_query['albums']['items'][0]['artists'][0]['name']
        art_uri = album_query['albums']['items'][0]['artists'][0]['uri']
    
    albums_uri_db['album_id'].append(alb_id)
    albums_uri_db['album'].append(alb_name)
    albums_uri_db['album_uri'].append(alb_uri)
    albums_uri_db['artist'].append(art_name)
    albums_uri_db['artist_uri'].append(art_uri)
    
    df_uri = pd.DataFrame.from_dict(albums_uri_db)
    
    return(df_uri)
```

```python
def find_tracks_uri(id_search, album_uri_search, artist_uri_search):
    tracks = sp.album_tracks(album_uri_search)
    album_specs = sp.album(album_id = album_uri_search)
    artist_specs = sp.artist(artist_id=artist_uri_search)
    spotify_albums = {'album_id':[],
                      'album_uri': [],
                      'album_type': [],
                      'album_popularity':[],
                      'album_genres':[],
                      'artist_genres':[],
                      'track_number': [],
                      'track_duration_ms': [],
                      'track_explicity': [],
                      'id': [],
                      'name': [],
                      'track_uri':[],
                      'acousticness' : [],
                      'danceability' : [],
                      'energy' : [],
                      'instrumentalness' : [],
                      'liveness' : [],
                      'loudness' : [],
                      'speechiness' : [],
                      'tempo' : [],
                      'valence' : []}

    for n in range(len(tracks['items'])):
        spotify_albums['album_id'].append(id_search)
        spotify_albums['album_uri'].append(album_uri_search)
        spotify_albums['album_type'].append(album_specs['type'])
        spotify_albums['album_popularity'].append(album_specs['popularity'])
        spotify_albums['album_genres'].append(album_specs['genres'])
        spotify_albums['artist_genres'].append(artist_specs['genres'])
        spotify_albums['track_number'].append(tracks['items'][n]['track_number'])
        spotify_albums['track_duration_ms'].append(tracks['items'][n]['duration_ms'])
        spotify_albums['track_explicity'].append(tracks['items'][n]['explicit'])
        spotify_albums['id'].append(tracks['items'][n]['id'])
        spotify_albums['name'].append(tracks['items'][n]['name'])
        spotify_albums['track_uri'].append(tracks['items'][n]['uri'])
        
    return(spotify_albums)
```

```python
def find_tracks_features(spotify_albums):
    for u in range(len(spotify_albums['track_uri'])):
        features = sp.audio_features(spotify_albums['track_uri'][u])
        if features[0] != None:
            spotify_albums['acousticness'].append(features[0]['acousticness'])
            spotify_albums['danceability'].append(features[0]['danceability'])
            spotify_albums['energy'].append(features[0]['energy'])
            spotify_albums['instrumentalness'].append(features[0]['instrumentalness'])
            spotify_albums['liveness'].append(features[0]['liveness'])
            spotify_albums['loudness'].append(features[0]['loudness'])
            spotify_albums['speechiness'].append(features[0]['speechiness'])
            spotify_albums['tempo'].append(features[0]['tempo'])
            spotify_albums['valence'].append(features[0]['valence'])
        else:
            spotify_albums['acousticness'].append('')
            spotify_albums['danceability'].append('')
            spotify_albums['energy'].append('')
            spotify_albums['instrumentalness'].append('')
            spotify_albums['liveness'].append('')
            spotify_albums['loudness'].append('')
            spotify_albums['speechiness'].append('')
            spotify_albums['tempo'].append('')
            spotify_albums['valence'].append('')
    
    df = pd.DataFrame.from_dict(spotify_albums)
    
    return(df)
```
